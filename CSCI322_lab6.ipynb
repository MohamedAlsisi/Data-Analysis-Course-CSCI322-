```python
## your code here
x_ = np.mean(X)
y_ = np.mean(Y)
a = sum([ (x - x_) * (y - y_) for x , y in zip (X , Y )]) / sum((x - x_) ** 2 for x in X)
b = y_ - a * x_
```

```python
a
```

```python
b
```

```python
yhat = lambda x: a * x + b
```

```python
plt.scatter(X,Y)
plt.plot([X.min(), X.max()], [yhat(X.min()), yhat(X.max())], color = 'g')
plt.scatter(X.mean(), Y.mean(), marker = 'x',  c = 'r' )
```

```python
## your code here
TSS = sum([(y - y_)** 2 for y in Y])
RSS = sum([(y - yhat(x))**2 for x , y in zip(X , Y)])
R2 = 1 - RSS/TSS
```
